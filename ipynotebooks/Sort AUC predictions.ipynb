Sort by one subject's score and list them all.

```python
nsub = 8
with open('auc_scores/AUC_scores.csv', 'rb') as csvfile:
    cv = csv.reader(csvfile, delimiter='\t')
    sortedlist = sorted(cv, key=operator.itemgetter(nsub), reverse=True)
for row in sortedlist:
    row[0].ljust(40)
    print('{0}\t{1}\t{2}\t{3}\t{4}\t{5}\t{6}\t{7}\t{8}'.format(\
                                                               row[0].ljust(35),\
                                                               row[1][:5],row[2][:5],row[3][:5],row[4][:5],\
                                                               row[5][:5],row[6][:5],row[7][:5],row[8][:5]))
```
