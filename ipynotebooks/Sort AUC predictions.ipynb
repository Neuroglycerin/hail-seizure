Loading these jsons, in order.

```python
discjsons = {}
for row in sortedlist:
    with open("discriminate_chain/"+row[0]+".json") as fh:
        discjsons[row[0]] = json.load(fh)
```

Iterate over stochastic optimisation results:

```python
import glob
import pdb
```

```python
opt_jsons = glob.glob("hdf5mcmc/*.json")
```

```python
overlap = []
for row in sortedlist[:3]:
    bestdisc  = discjsons[row[0]]
    for fname in opt_jsons:
        with open(fname) as fh:
            optsettings = json.load(fh)
        # then compare the set of features
        overlap_value = 1 - (len(set(bestdisc['FEATURES'])-set(optsettings['FEATURES'])) + \
                                len(set(optsettings['FEATURES']) - set(bestdisc['FEATURES']))/ \
                                (len(set(bestdisc['FEATURES'])) + len(set(optsettings['FEATURES'])) ))
        overlap.append((row[0],fname,overlap_value))
        if overlap_value > 0.9999:
            pdb.set_trace()
```

Ok, now we want to know what score these ones got when they were run:

```python
overlaphashes = list(map(lambda x: x.split(".")[0].split("/")[1], list(zip(*overlap))[1]))[:10]
```

```python
scores = {}
with open("hdf5mcmc/AUC_scores.csv") as fh:
    c = csv.reader(fh, delimiter="\t")
    fline = next(c)
    for row in c:
        if row[0] in overlaphashes:
            scores[row[0]] = row[1:]
```
