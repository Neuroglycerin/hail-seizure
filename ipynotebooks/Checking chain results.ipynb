```python
import glob
import json
```

```python
jsons = glob.glob("*.json")
```

```python
jsons
```

```python
results = {}
for fname in jsons:
    with open(fname) as fh:
        results[fname] = json.load(fh)
```

Now loaded all the settings of all the results, will iterate through and compare all of the features arrays in each settings files to all the others.

```python
import itertools
```

```python
pairs = []
for json1,json2 in itertools.combinations(jsons,2):
    if set(results[json1]['FEATURES']) == set(results[json2]['FEATURES']):
        print("{0} contains same features as {1}".format(json1,json2))
        pairs.append((json1,json2))
```

Looks like it's not detecting when it has the same features properly.
Looking some of these more closely:

```python
print(results[pairs[0][0]]['FEATURES'])
print(results[pairs[0][1]]['FEATURES'])
```

So what's happening is these are all cases where it has selected the same feature twice.
That shouldn't be allowed.

Checking this is what's happening in the other cases:

```python
for pair in pairs:
    print(set(results[pair[0]]['FEATURES']),set(results[pair[1]]['FEATURES']))
```

Yeah, looks like that's what's happened.
